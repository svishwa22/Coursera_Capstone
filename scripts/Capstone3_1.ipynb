```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
df_list = pd.read_html(url)
```

```python
#no. of tables
len(df_list)
```

```python
#let's check first table
df_list[0]
```

```python
# as this is the relevant table let's get this in a dataframe
# Get first table                                                                                                           
df = df_list[0]
df.head()
```

```python
df.Borough.unique()
```

```python
#removing rows where Borough is 'Not Assigned' and resetting index
df1=df[df['Borough'] != 'Not assigned'].reset_index(drop=True)
df1.head()
```

```python
df1.Borough.unique()
```

```python
# check if there is any Neighbourhood with 'Not assigned'
df2=df1[df1['Neighbourhood'] == 'Not assigned']
df2.head()
```

```python
# as there is no Neighbourhood with 'Not assigned', we can use df1 for further analysis
print('number of rows of our dataframe:' ,df1.shape)
```
